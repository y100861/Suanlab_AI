## 일차함수
* $y = ax + b$
  + $a$: 기울기, $b$: y절편
* 그래프 상에서 직선인 그래프(linear)

```python
def linear_function(x):
    a = 0.5
    b = 2
    return a*x + b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5, 5, 0.1)
y = linear_function(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Function")
plt.show()
```

## 이차함수
* $y = ax^2 + bx + c$
* 일반적으로 두 개의 실근을 가짐

```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    return a*x**2 + b*x + c
```

```python
print(quadratic_function(2))
```

```python
x = np.arange(-5, 5, 0.1)
y = quadratic_function(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Function")
plt.show()
```

## 삼차함수(다항함수)  

* $y = ax^3 + bx^2 + cx + d$

```python
def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8
    return a*x**3 + b*x**2 + c*x + d
```

```python
print(cubic_function(3))
```

```python
x = np.arange(-5, 5, 0.1)
y = cubic_function(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Cubic Function")
plt.show()
```

### 함수의 최소값/최대값


```python
def my_func(x):
    a = 1
    b = -3
    c = 10
    return a*x**2 + b*x + c
```

```python
x = np.arange(-10, 10, 0.1)
y = my_func(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(1.5, my_func(1.5))
plt.text(1.5-1.5, my_func(1.5)+10, "min value of f(x)\n({}, {})".format(1.5, my_func(1.5)), fontsize=10)
plt.title("my_func")
plt.show()
```

```python
min_val = min(y)
print(min_val)
```

### 특정 구간 내에서 최소값 구하기

```python
def get_minimum(x1, x2, f):
    x = np.arange(x1, x2, 0.01)
    y = f(x)

    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("get_minimum")
    plt.show()

    return min(y)
```

```python
print(get_minimum(1, 4, my_func))
```

## 지수함수

- $y = a^x$ ($a \ne 0$) (기본형)
- $y = e^x$ ($e = 2.71828...$)

```python
def exponential_function(x):
    a = 4
    return a**x
```

```python
print(exponential_function(4))
print(exponential_function(0))
```

```python
x = np.arange(-3, 2, 0.1)
y = exponential_function(x)

plt.plot(x, y)
plt.xlabel("x")
plt.xlabel("y")
plt.title("exponential_function")
plt.show()
```

```python
def exponential_function2(x):
    a = 3
    return math.pow(a, x)
```

## 극한
- 극한에 대해서는 어떠한 식을 코드로 표현할 수 있다 정도로만 이해하며 참고
- 극한에서 알아야 할 사실은 x가 어떤 값 a에 가까이 다가갈 때 a에 '한없이 가까이 간다'일 뿐, a에 도달하지 않는다는 점
- 이를 표현할 때, 엡실론(epsilon)이라는 아주 작은 값(ex, 0.0001) 등으로 표현


```python
from sympy import *
init_printing()

x, y, z = symbols("x y z")
a, b, c, t = symbols("a b c t")
```

- $\lim_{x \to 1} \left(\frac{x^3-1}{x-1} \right) = 3$

```python
print("극한값: {}".format(limit((x**3-1) / (x-1), x, 1)))
plot(((x**3-1) / (x-1)), xlim=[-5, 5], ylim=[-1, 10]);
```

- $\lim_{x \to \infty} \left(\frac{1+x}{x} \right)$

```python
print("극한값: {}".format(limit(((1+x) / x), x, oo)))
plot(((1+x) / x), xlim=[-10, 10], ylim=[-5, 5]);
```

- $\lim_{x \to 1}\frac{\sqrt{x+3}-2}{x-1} = \frac {1}{4}$

```python
print("극한값: {}".format(limit((sqrt(x+3) - 2) / (x-1), x, 1)))
plot((sqrt(x+3) - 2) / (x-1), xlim=[-5, 12], ylim=[-0.5, 1]);
```

### 삼각함수의 극한

- $\lim_{x \to \frac{\pi}{2} +0}\ \tan x = -\infty$
- $\lim_{x \to \frac{\pi}{2} -0}\ \tan x = \infty$

```python
print("극한값: {}".format(limit(tan(x), x, pi/2, "+")))
print("극한값: {}".format(limit(tan(x), x, pi/2, "-")))
plot(tan(x), xlim=[-5, 5], ylim=[-10, 10])
```

- $\lim_{x \to 0} \left(\frac{\sin x}{x} \right) = 1$

```python
print("극한값: {}".format(limit(sin(x) / x, x, 0)))
plot((sin(x) / x));
```

- $\lim_{x \to 0} x \sin \left(\frac{1}{x} \right)$

```python
print("극한값: {}".format(limit(x * sin(1/x), x, 0)))
plot((x * sin(1/x)), xlim=[-2, 2], ylim=[-1, 1.5]);
```

### 지수함수, 로그함수의 극한

### 자연로그(e)의 밑
(1) $lim_{x \to \infty} \ (1 + \frac{1}{x})^x \ = e$  
(2) $\lim_{x \to \infty} \ (1 + \frac{2}{x}^x) \ = e^2$  
(3) $\lim_{x \to 0} \ \frac{(e^x - 1)}{x} = 1$  
(4) $\lim_{x \to 0} \ \frac{ln(1 + x)}{x} = 1$

```python
print("(1) 극한값: {}".format(limit((1 + 1/x)**x, x, oo)))
print("(2) 극한값: {}".format(limit((1 + 2/x)**x, x, oo)))
print("(3) 극한값: {}".format(limit((exp(x)-1) / x, x, 0)))
print("(4) 극한값: {}".format(limit(ln(1+x) / x, x, 0)))

plot(ln(1+x) / x, xlim=[-4, 6], ylim=[-2, 8])
```

## 미분
- 어떤 한 순간의 변화량을 표시한 것

### 미분과 기울기

- 어떤 함수를 나타내는 그래프에서 한 점의 미분값(미분계수)를 구하는 것은 해당 점에서의 접선을 의미

![](https://upload.wikimedia.org/wikipedia/commons/thumb/0/0f/Tangent_to_a_curve.svg/440px-Tangent_to_a_curve.svg.png)
<br /><sub>출처: https://ko.wikipedia.org/wiki/%EB%AF%B8%EB%B6%84</sub>

* 기울기는 **방향성**을 가짐
  + 이용할 미분 식 (수치 미분)  
      $\frac{df(x)}{dx} = \lim_{x \to \infty} \frac{f(x+h) - f(x-h)}{2h}$

* [주의] $ \ h \ $는 아주 작은 수를 뜻하는데, 예를 들어 $\ 10e-50 \ $ 정도의 수를 하면 파이썬은 이를 $0.0$으로 인식
* 따라서, 딥러닝에서 아주 작은 수를 정할 때 $\ 1e-4\ $ 정도로 설정해도 무방

```python
# 미분 식
def numeric_differential(f, x):
    h = 1e-4
    return (f(x+h) - f(x-h)) / (2*h)
```

### 함수 위의 점 $(a, b)$에서의 접선의 방정식

  - 예제 : 점 (1, 7) 에서의 기울기


```python
def my_func(x):
    return 2*x**2 + 3*x + 2
```

```python
def linear_function(a, b, c, x):
    return c*(x-a) + b
```

```python
c = numeric_differential(my_func, 1)

x = np.arange(-5, 5, 0.01)
y = linear_function(1, my_func(1), c, x)

plt.scatter(1, my_func(1))
plt.plot(x, my_func(x), x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x) & linear function")
plt.show()
```

### 미분 공식
- $\frac{d}{dx} (c) = 0\ (c는 상수)$

- $\frac{d}{dx} [cf(x)] = cf^\prime (x)$

- $\frac{d}{dx} [f(x) + g(x)] = f^\prime(x) + g^\prime(x)$

- $\frac{d}{dx} [f(x)-g(x)] = f^\prime(x)\ - \ g^\prime(x)$

- $\frac{d}{dx} [f(x)g(x)] = f(x)g^\prime(x)\ + \ f^\prime(x)g(x)\ (곱셈공식)$

- $\frac{d}{dx} [\frac{f(x)}{g(x)}] = \frac{g(x)f^\prime(x)\ - \ f(x)g^\prime(x)}{[g(x)]^2}$

- $\frac{d}{dx} [x^n] = nx^n-1$

### 편미분
- 변수가 1개짜리인 위의 수치미분과 달리, 변수가 2개 이상일 때의 미분법을 편미분이라 함
- 다변수 함수에서 특정 변수를 제외한 나머지 변수는 상수로 처리하여 미분을 하는 것
- 각 변수에 대해 미분 표시를 $\sigma$를 통해 나타남
- ex) $f(x_0, x_1) = x_0^2 + x_1^2$  


```python
def my_func2(x):
    return (x[0]**2 + x[1]**2)
```

```python
x = np.arange(-4.0, 4.0, 0.4)
y = np.arange(-4.0, 4.0, 0.4)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.figure(figsize=(10, 7))
ax = plt.subplot(projection="3d")
ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm);
plt.title("f(x, y) = x**2 + y**2", fontdict={"weight": "bold", "size": 15})
plt.show()
```

- 예제1 : x0에 대한 편미분, $\frac{\sigma f}{\sigma x_0}$

```python
x = np.array([1, 2])

def f0_function(x0):
    return x0**2 + 2**2

print("미분값: {}".format(numeric_differential(f0_function, x[0])))
```

- 예제2 : x1에 대한 편미분, $\frac{\sigma f}{\sigma x_1}$

```python
def f1_function(x1):
    return 1**2 + x1**2

print("미분값: {}".format(numeric_differential(f1_function, x[1])))
```

### 기울기(gradient)
- 방향성을 가짐

```python
def numeric_diff(f, x):
    h = 1e-5
    grad = np.zeros_like(x)
    
    for i in range(x.shape[0]):
        k = x[i]

        x[i] = k + h
        fx1 = f(x)

        x[i] = k - h
        fx2 = f(x)

        x[i] = k
        grad[i] = (fx1 - fx2) / (2*h)

    return grad

print(numeric_diff(my_func2, np.array([1.0, 2.0])))
```

### 기울기의 의미를 그래프로 확인
* 기울기가 가장 낮은 장소(가운데)로부터 거리가 멀어질수록 기울기가 커짐
* 기울기가 커진다는 것은 영향을 많이 받는다는 의미  
* 기울기가 작다는 것은 영향을 적게 받는다는 의미
